```python
import pandas as pd
import wbgapi as wb
```

```python
#Pandas function has been used to read the values of the dataset from the indicators fetched from World Indicators Repository
dfr1=pd.read_csv(r"C:\\Sai Teja\Applied Data Science 2\World Bank Data Final.csv", low_memory=False)
```

```python
#Initialising the indicators both related to economy and climate
indi_GD = ['SL.UEM.ADVN.ZS','NE.DAB.TOTL.ZS','NE.IMP.GNFS.ZS','NY.GDP.MKTP.PP.CD']
conty_cde = ["AFG","IND",'AUS','PAK','BGR','GBR','ESP','LUX','CHL','CHE']
indi_CL=['EN.ATM.PM25.MC.T1.ZS','EN.ATM.CO2E.GF.KT','ER.LND.PTLD.ZS','EG.ELC.ACCS.RU.ZS']
G  = wb.data.DataFrame(indi_GD, conty_cde, mrv=7)
C = wb.data.DataFrame(indi_CL, conty_cde, mrv=7)
#SL.UEM.ADVN.ZS: Unemployment with advanced education (% of total labor force with advanced education)	
#NE.DAB.TOTL.ZS: Expenditure of the country in total
#NE.IMP.GNFS.ZS: Import of the country in total
#NY.GDP.MKTP.PP.CD: PPP, GDP of a country
#EN.ATM.PM25.MC.T1.ZS: PM2.5 pollution exceeding WHO target levels
#EN.ATM.CO2E.GF.KT: CO2 emissions from fuel consumptions of gas 
#ER.LND.PTLD.ZS	Terrestrial protected areas (% of total land area)	
#EG.ELC.ACCS.RU.ZS: Rural electricity access
```

```python
# Economy indicators
G.columns = [s.replace('YR','') for s in G.columns]      
G=G.stack().unstack(level=1)                             
G.index.names = ['Cnt_Cde', 'Year']                           
G.columns                                                     
G.fillna(0)
G.head(10)
```

```python
# Cllimate indicators
C.columns = [s.replace('YR','') for s in C.columns]      
C=C.stack().unstack(level=1)                             
C.index.names = ['Cnt_Cde', 'Year']                           
C.columns                                                     
C.fillna(0)
C.head(10)
```

```python
#Dataset prep
dp1=G.reset_index()
dp2=C.reset_index()
dp3=dp1.fillna(0)
dp4=dp2.fillna(0)
```

```python
#Merging the values
dpfinal = pd.merge(dp3, dp4)
dpfinal.head(14)
```

```python
# Summary of Afghanistan 
su1=dpfinal[(dpfinal['Cnt_Cde']=='AFG')]
su1.describe()
```
